```python
mvp_odds = pd.read_csv('mvp_odds.csv').rename(columns={'Unnamed: 0':'passer'})

filter_ = (qbs.season==2021)

market_vs_production = qbs.loc[filter_]\
    [['season','passer','epa','bayesian_career_epa','bayesian_ewm','bayesian_career_blend','bayesian_career_blend_percentile']]\
    .sort_values('bayesian_career_blend', ascending=False)\
    .merge(mvp_odds[['passer','odds','juice adj implied prob']].rename(columns={'odds':'MVP odds','juice adj implied prob':'MVP probability'}), on='passer')

market_vs_production['residual'] = helpers.residual(market_vs_production, 'bayesian_career_blend', 'MVP probability')
market_vs_production['abs_residual'] = abs(market_vs_production['residual'])

market_vs_production\
    .sort_values('residual', ascending=False)\
    .style.background_gradient(subset=['bayesian_career_blend','MVP probability', 'residual'], cmap=cm.get_cmap('PRGn'))\
    .background_gradient(subset=['MVP odds'], cmap=cm.get_cmap('PRGn_r'))

market_vs_production\
    .loc[market_vs_production['abs_residual']<.01]\
    [['passer','bayesian_career_blend','bayesian_career_blend_percentile', 'MVP odds', 'MVP probability', 'residual']]\
    .sort_values('MVP probability', ascending=False)\
    .style\
    .background_gradient(subset=['bayesian_career_blend'], cmap=cm.get_cmap('PRGn'), vmin=market_vs_production['bayesian_career_blend'].min(), vmax=market_vs_production['bayesian_career_blend'].max())\
    .background_gradient(subset=['MVP odds'], cmap=cm.get_cmap('PRGn_r'), vmin=market_vs_production['MVP odds'].min(), vmax=market_vs_production['MVP odds'].max())\
    .background_gradient(subset=['MVP probability'], cmap=cm.get_cmap('PRGn'), vmin=market_vs_production['MVP probability'].min(), vmax=market_vs_production['MVP probability'].max())
```
